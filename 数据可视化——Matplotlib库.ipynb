# 数据可视化——Matplotlib库

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy as sp
```

## 1 使用pyplot创建图形

### 1.1 创建简单的图形

```python
# 创建figure对象的语法格式
# plt.figure(num=None, figsize=None, 
# dpi=None, facecolor=None, 
# edgecolor=None, frameon=True,
# FigureClass=<class 'matplotlib.figure.Figure'>,
# clear=False, **kwargs)
# 参数说明：
# num:图像编号或名称，数字为编号 ，字符串为名称
# figsize:指定figure的宽和高，单位为英寸；
# dpi参数指定绘图对象的分辨率，即每英寸多少个像素，缺省值为80      1英寸等于2.5cm,A4纸是 21*30cm的纸张
# facecolor:背景颜色
# edgecolor:边框颜色
# frameon:是否显示边框
# FigureClass:选择图像类型，如：plt.figure(FigureClass=fig.Figure)
# clear:清空当前图像，默认为False
# **kwargs:其他参数

# 绘制图表
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1,
         label = 'sin(x)',
         color = 'red',
         linewidth = 1.0,
         linestyle = '-')
plt.plot(x, y2,
         label = 'cos(x)',
         color = 'green',
         linewidth = 1.0,
         linestyle = '-.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x) & cos(x)')
plt.legend(loc = 'upper left')# 图例
plt.show()
```

### 1.2 创建子图

```python
# 创建子图
# plt.subplot(nrows, ncols, plot_number)
k = x
plt.subplot(2, 2, 1)
plt.plot(x, y1,
         label='sin(x)',
         color='red',
         linewidth=1.0,
         linestyle='-')
plt.legend(loc='upper left')  # 图例
plt.title('sin(x)')
plt.ylabel('y')
plt.subplot(222)
plt.plot(x, y2,
         label='cos(x)',
         color='green',
         linewidth=1.0,
         linestyle='-.')
plt.title('cos(x)')
plt.legend(loc='upper left')  # 图例
plt.ylabel('y')
plt.subplot(212)
plt.plot(x, k,
         label='x',
         color='black',
         linewidth=1.0,
         linestyle='dotted')
plt.legend(loc='upper left')  # 图例
plt.xlabel('x')
plt.ylabel('k')
plt.title('(x)')
plt.show()
```

## 2 分析变量间关系图

### 2.1 绘制散点图

```python
plt.scatter(x, y1, marker='.',
            color='red', s=40,
            label='sin(x)')
plt.scatter(x, y2, marker='.',  # 点的形状
            color='green', s=40,
            label='cos(x)')
plt.legend(loc='upper left')  # 图例
plt.xlabel('x')
plt.ylabel('y')
del x, y1, y2, k
```

### 2.2 绘制折线图

```python
x = np.arange(0, 10, 0.1)
y1 = np.concatenate((np.arange(0, 5, 0.1),
                     np.arange(5, 0, -0.1)))
y2 = np.concatenate((np.arange(0, 5, 0.1)*2,
                     np.arange(5, 0, -0.1)*2))
y3 = np.concatenate((np.arange(0, 5, 0.1)**0.5,
                     np.arange(5, 0, -0.1)**0.5))
plt.plot(x, y1,
         label='test1',
         linestyle='-')  # 线条类型
plt.plot(x, y2,
         label='test2',
         linestyle='--')  # 线条类型
plt.plot(x, y3,
         label='test3',
         linestyle='-.')  # 线条类型
plt.legend(loc='upper left')  # 图例
plt.show()
del x, y1, y2, y3
```

## 3 分析变量数据分布和分散状况

### 3.1 绘制直方图

```python
np.random.seed(2023)
x = np.random.normal(0, 1, 1000)
plt.hist(x, bins=10, color='green',
            alpha=0.5, rwidth=0.9,
            orientation='vertical')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram')
# 参数说明
# x:指定要绘制直方图的数据；
# bins:指定直方图条形的个数；
# color:指定直方图的填充色；
# alpha:指定直方图的透明度；
# rwidth:指定直方图条形的宽度，也就是直方图整体宽度的占比；
# histtype:指定直方图的类型，默认为bar，除此还有’barstacked’, ‘step’,  ‘stepfilled’；
# label:设置直方图的标签，可通过legend展示其图例；
# stacked:当有多个数据时，是否将直方图呈堆叠摆放，默认水平摆放；
# orientation:设置直方图的摆放方向，默认为垂直方向；
# log:是否对频率取对数。
plt.show()
del x
```

### 3.2 绘制柱状图

```python
# 绘制柱状图示例
x = np.arange(0, 10, 1)
y1 = np.concatenate((np.arange(0, 5, 1),
                        np.arange(5, 0, -1)))
y2 = np.concatenate((np.arange(0, 5, 1)*2,
                        np.arange(5, 0, -1)*2))
y3 = np.concatenate((np.arange(0, 5, 1)**0.5,
                        np.arange(5, 0, -1)**0.5))
plt.bar(x-0.2, y1,
        label='test1',
        width=0.15,  # 柱状图的宽度
        color='green',
        alpha=1)  # 透明度
plt.bar(x, y2,
        label='test2',
        width=0.15,  # 柱状图的宽度
        color='red',
        alpha=0.5)  # 透明度
plt.bar(x+0.2, y3,
        label='test2',
        width=0.15,  # 柱状图的宽度
        color='blue',
        alpha=0.5)  # 透明度
plt.legend(loc='upper left')  # 图例
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar')
plt.show()

del x, y1, y2, y3
```

### 3.3 绘制饼图

```python
# 绘制饼图示例
x = np.array([1, 2, 3, 4, 5])
plt.pie(x,
        labels=['A', 'B', 'C', 'D', 'E'],
        colors=['red', 'green', 'blue', 'yellow', 'pink'],
        autopct='%.2f%%',  # 数值保留固定小数位
        explode=[0, 0.05, 0, 0, 0],  # 每一块离开中心距离
        startangle=0,  # 第一块起始角度
        shadow=True)  # 阴影
plt.title('Pie')
plt.legend(loc='upper left')  # 图例
plt.show()
```

### 3.4 绘制箱线图

```python
# 绘制箱线图示例
np.random.seed(2023)
x = [np.random.normal(0, 1, 1000),
        np.random.normal(1, 1, 1000),
        np.random.normal(2, 1, 1000)]
label = ['A', 'B', 'C'] 
plt.boxplot(x,
            labels = label,
            notch=True,  # 中间箱体是否缺口
            sym='o',  # 异常点形状
            vert=True,  # 是否垂直
            whis=1.5,  # IQR
            patch_artist=False,  # 上下四分位框是否填充
            meanline=False,  # 是否有均值线及其形状
            showmeans=True,  # 是否显示均值
            showcaps=True,  # 是否显示箱线图顶端和末端的两条线
            showbox=True,  # 是否显示箱线图的箱体
            showfliers=True,  # 是否显示异常值
            widths=0.5,  # 箱体宽度
            )
plt.show()
```
